VISUALIZATION

1.SALES DISTRIBUTION BY STATE

```python
df.groupby('State')['Sales'].sum().sort_values().plot(kind='bar', figsize=(10,6), color='skyblue')
plt.xlabel('State')
plt.ylabel('Sales')
plt.title('Sales distribution by States')
plt.tight_layout()
plt.show()
```

```python
# Aggregate sales by state
state_sales = df.groupby('State')['Sales'].sum().reset_index()

# Create bar plot
plt.figure(figsize=(12,10))
sns.barplot(data=state_sales, x='State', y='Sales', palette='viridis')
plt.title('Total Sales by State')
plt.xlabel('Sales')
plt.xticks(rotation=90)
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

2.SALES BY CATEGORY ACROSS STATE

```python
# Aggregate sales by State and Category
grouped = df.groupby(['State', 'Category'])['Sales'].sum().reset_index()

# Plot
plt.figure(figsize=(14,8))
sns.barplot(data=grouped, x='State', y='Sales', hue='Category')
plt.xticks(rotation=90)
plt.title('Sales by Category across States')
plt.tight_layout()
plt.show()
```

3.SALES BY SUB-CATEGORY ACROSS STATE

```python
# Aggregate sales
grouped_sub = df.groupby(['State', 'Sub-Category'])['Sales'].sum().reset_index()

# Plot
plt.figure(figsize=(20,8))
sns.barplot(data=grouped_sub, x='State', y='Sales', hue='Sub-Category')
plt.xticks(rotation=90)
plt.title('Sales by Sub-Category across States')
plt.tight_layout()
plt.show()
```

```python
grouped_sub1 = df.groupby(['State', 'Sub-Category'])['Sales'].sum().reset_index()

g = sns.FacetGrid(grouped_sub1, col='Sub-Category', col_wrap=4, height=5, sharey=False)
g.map_dataframe(sns.barplot, x='State', y='Sales', order=grouped_sub1['State'].unique())
g.set_xticklabels(rotation=90)
g.fig.suptitle('Sales by State for Each Sub-Category', fontsize=20,y=1.01)
plt.tight_layout()
plt.show()
```

4.DISTRIBUTION OF (TECHNOLGOY)SUB-CATEGORY BY SALES

```python
# Filter for Technology category
tech_df = df[df['Category'] == 'Technology']

# Group by Sub-Category 
grouped = tech_df.groupby(['Sub-Category'])['Sales'].sum().reset_index()

# Sort and plot top products
top_products = grouped.sort_values(by='Sales', ascending=False).head(20)


plt.figure(figsize=(16,8))
sns.barplot(data=top_products, x='Sub-Category', y='Sales', hue='Sub-Category')
plt.title('Top Technology Products by Sales')
plt.tight_layout()
plt.show()
```
